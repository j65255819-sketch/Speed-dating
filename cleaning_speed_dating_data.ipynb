```python
import zipfile
import pandas as pd

# Unzip the file
with zipfile.ZipFile('/content/Speed Dating Data.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/')

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/Speed Dating Data.csv', encoding='latin1')

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
print(df[['iid', 'pid', 'race', 'race_o','dec']].isnull().sum())
```

```python
print(f"There are {df.shape[0]} rows in the DataFrame.")
```

```python
initial_rows = df.shape[0]
df.dropna(subset=['pid', 'race', 'race_o'], inplace=True)
rows_dropped = initial_rows - df.shape[0]
print(f"Dropped {rows_dropped} rows due to missing values in 'pid', 'race', or 'race_o'.")
print(f"New DataFrame shape: {df.shape}")
```

```python
output_filename = 'Speed Dating Data_cleaned.csv'
df.to_csv(output_filename, index=False)
print(f'DataFrame successfully saved to {output_filename}')
```
